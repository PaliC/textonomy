```python
import pandas as pd

data = pd.read_csv('search_strings.csv', encoding = "ISO-8859-1")
data['item_title'] = data['item_title'].str.lower()
```

```python
count = 0
descset = set()
dupeset = set()

for _, val in data['item_title'].iteritems():
    if val in descset:
        count = count + 1
        dupeset.add(val)
    else:
        descset.add(val)
        
```

```python
print(dupeset)
print(len(dupeset))
print(count)
```
